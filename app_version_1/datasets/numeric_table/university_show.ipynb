```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('university_profile_raw.csv')
```

```python
df.head()
```

```python
df.columns = [
    'University',
    'Total publications per year',
    'Community size',
    'Number of publications per professor per year',
    'Average impact factor',
    'link'
]
```
